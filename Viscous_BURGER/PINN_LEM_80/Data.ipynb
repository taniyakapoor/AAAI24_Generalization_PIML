```python
import scipy.io
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.tri import Triangulation

# Load MATLAB file
mat_data = scipy.io.loadmat('cylinder_nektar_t0_vorticity.mat')
```

```python
mat_data
```

```python
# Access variables in the MATLAB file
x = mat_data['x'].reshape(-1)
y = mat_data['y'].reshape(-1)
w = mat_data['w'].reshape(-1)

w.shape
```

```python
# Create a triangulation of the (x, y) points
triangulation = Triangulation(x, y)
plt.figure(figsize=(16, 9))
# Plot the contour plot
plt.tricontourf(triangulation, w, levels=998, cmap='seismic')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.colorbar(label='Z-axis')
plt.show()
```
